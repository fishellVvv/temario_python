## Funciones

```python
def saludar():
    print("Hola!")

saludar()
```

- Reutilizar código
- Almacenar bloques de ejecución
- Ordenar nuestro código
- Mantener mejor nuestro código al estar separado por funciones

```python
def saludar_nombre(nombre):
    print(f"Hola {nombre}!")

saludar_nombre("Antonio")
```

```python
def sumar_dos_numeros(numA, numB):
    total = numA + numB
    print(total)

sumar_dos_numeros(5, 8)
```

```python
def multiplicar_por_dos(numero):
    total = numero * 2
    return total

print(multiplicar_por_dos(21))
```

```python
def saludar_usuario(nombre_usuario = "User"):
    print(f"Hola {nombre_usuario}")

saludar_usuario()
saludar_usuario("Victor")
```

```python
x = 10

def mostrar_valor():
    x = 5
    print("Valor dentro de la función", x)

mostrar_valor()
print("Valor fuera de la función", x)
```

```python
def sumar_numeros(numA, numB):
    """
    Suma dos números y devuelve el resultado

    Args:
        numA (int): el primer número de la suma
        numB (int): el segundo número de la suma

    Returns:
        int: El resultado de la suma de ambos números
    """
    return(numA + numB)

print(sumar_numeros(5, 8))
```
